```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
norm.pdf(0)
```

```python
x = np.linspace(-4, 4, 100)
```

```python
y_pdf = norm.pdf(x)  # 標準正規分布
```

```python
plt.plot(x,y_pdf)
plt.xlabel('x')
plt.ylabel('p')
plt.title('Standard Normal Distribution')
pass
```

```python
y_cdf = norm.cdf(x)  # 標準正規分布
```

```python
plt.plot(x, y_cdf)
plt.xlabel('x')
plt.ylabel('P')
plt.title('Cumulative Distribution Function')
pass
```

```python
norm.cdf(0)
```

```python
norm.cdf(-4)
```

```python
1-norm.cdf(4)
```

```python
p = np.linspace(0,1,100)
y_ppf = norm.ppf(p)  # 標準正規分布
```

```python
plt.plot(p,y_ppf)
plt.xlabel('P')
plt.ylabel('x')
plt.title('Percent Point Function')
pass
```

```python
norm.ppf(0.5)
```

```python
norm.ppf(0.025)
```

```python
norm.ppf(0.975)
```

```python
y_rvs = norm.rvs(size=10_000)  # 標準正規分布
```

```python
plt.hist(y_rvs, bins=30)  #  bins=表示する棒の数（デフォルトは10)
pass
```

```python
from scipy.stats import gaussian_kde  # サブパッケージを読み込む
kde = gaussian_kde(y_rvs)  # y_rvsから確率密度関数を推定
plt.plot(x, kde(x))        # 横軸のxに対してプロット
pass
```

```python
plt.hist(y_rvs, bins=30, density=True)
plt.plot(x, kde(x))
pass
```

```python
from scipy.stats import t
x = np.linspace(-4,4,100)
y = t.pdf(x, df=1)
plt.plot(x,y)
pass
```

```python
t.cdf(-3, df=1)
```

```python
1-t.cdf(3, df=1)
```

```python
from scipy.stats import chi2
x = np.linspace(0,12,100)
y = chi2.pdf(x, df=3)
plt.plot(x,y)
pass
```

```python
chi2.cdf(1, df=3)
```

```python
1-chi2.cdf(10, df=3)
```

```python
from scipy.stats import f
x = np.linspace(0.001,5,1000)
y = f.pdf(x, dfn=5, dfd=1)
plt.plot(x,y)
pass
```

```python
f.cdf(0.1, dfn=5, dfd=1)
```

```python
1-f.cdf(5, dfn=5, dfd=1)
```

```python
from scipy.stats import uniform
x = np.linspace(0,12,100)
y = uniform.pdf(x, loc=1, scale=9)
plt.plot(x,y)
pass
```

```python
from scipy.stats import logistic
x = np.linspace(-5,5,100)
y = logistic.pdf(x)
plt.plot(x,y)
pass
```
